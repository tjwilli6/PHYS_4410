```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("/Users/tjwilli/jupyter.mplstyle")
```

```python
x = np.linspace(0.5,5,1000)
n_mb = np.exp(-x)
n_be = 1 / (np.exp(x) - 1)
n_fd = 1 / (np.exp(x) + 1)
```

```python
plt.figure(figsize=(6,3))
plt.plot(x,n_mb,c='red',label='Maxwell-Boltmann')
plt.plot(x,n_be,c='blue',label='Bose-Einstein')
plt.plot(x,n_fd,c='green',label='Fermi-Dirac')
plt.axhline(0,c='k')
plt.xlabel(r'$\frac{\epsilon_s-\mu}{kT}$')
plt.ylabel(r'$\left<n_s\right>$')
plt.legend()
```

```python
energy_levels = []
nmax = 5
for nx in range(1,nmax):
    for ny in range(1,nmax):
        for nz in range(1,nmax):
            energy_levels.append(nx**2 + ny**2 + nz**2)
energy_levels = sorted(list(set(energy_levels)))
```

```python
def get_states_energy(nlevel=1):
    energy = energy_levels[nlevel-1]
    nmax = int(np.ceil( np.sqrt(energy) ) )
    states = []
    for nx in range(1,nmax+1):
        for ny in range(1,nmax+1):
            for nz in range(1,nmax+1):
                n = np.array([nx,ny,nz])
                if np.isclose(np.sum(n**2),energy):
                    states.append(n)
    return states
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(8,5))

ax = plt.axes(projection='3d')
n = 0
for i in range(len(energy_levels)):
    states = get_states_energy(i+1)
    z = 0
    for s in states:
        nx,ny,nz = s
        artist=ax.scatter(nx,ny,nz,color='C{}'.format(i))
        ax.set_xlabel(r'$n_x$')
        ax.set_ylabel(r'$n_y$')
        ax.set_zlabel(r'$n_z$')
        ax.view_init(45,45)
        ax.set_xlim(0,6)
        ax.set_ylim(0,6)
        ax.set_zlim(0,6)
        label = r'$n_x^2+n_y^2+n_z^2={}$'.format(energy_levels[i]) + '\n' + r'$n_x={},n_y={},n_z={}$'.format(nx,ny,nz)
        a = plt.legend([artist],[label])
        plt.savefig('movie/frame{}.pdf'.format(n),bbox_inches='tight')
        n+=1
#ax.view_init(45,45)
```
