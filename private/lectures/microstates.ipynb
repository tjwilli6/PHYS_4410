```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("/Users/tjwilli/jupyter.mplstyle")
```

```python
Na = 3000
Nb = 2000
qtot = 1000
```

```python
def get_num_microstates(N,q):
    fact = np.math.factorial
    return np.float128(fact(q + N - 1)) / np.float128(fact(q)) / np.float128(fact(N-1))
```

```python
qa_list = []
wtot_list = []
print("qA\tWa\t\tqB\tWb\t\tWtot")
print('-'*60)
for q in range(qtot):
    qa = q
    qa_list.append(qa)
    qb = qtot - qa
    num_states_a = get_num_microstates(Na,qa)
    num_states_b = get_num_microstates(Nb,qb)
    num_states_tot = num_states_a * num_states_b
    wtot_list.append(num_states_tot)
    print('{}\t{:.1E}\t\t{}\t{:.1E}\t\t{:.1E}'.format(qa,num_states_a,qb,num_states_b,num_states_tot))
```

```python
plt.plot(qa_list,wtot_list)
plt.xlabel(r'$q_a$')
plt.ylabel(r'$\Omega$')
```
