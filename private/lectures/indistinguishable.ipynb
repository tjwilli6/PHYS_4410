```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use("/Users/tjwilli/jupyter.mplstyle")
```

```python
n_levels = 5
n_particles = 2

rows = []
colors = []
i=0
for nA in range(n_levels):
    for nB in range(n_levels):
        row = np.empty(n_levels+1,dtype="U2")
        row.fill("")
        row[0] = str(i+1)
        row[nA+1] += "A"
        row[nB+1] += "B"
        colors.append( ['gray'] + ['white'] * 5 )
        rows.append(row)
        
        i+=1
```

```python
plt.table(rows,cellColours=colors)
plt.axis(False)
```

```python
n_levels = 5
n_particles = 2

rows = []
colors = []
i=0
for nA in range(n_levels):
    for nB in range(n_levels):
        row = np.zeros(n_levels+1,dtype=int)
        row[0] = i+1
        row[nA+1] += 1
        row[nB+1] += 1
        rows.append(row)
        colors.append( ['gray'] + ['white'] * 5 )
        i+=1
```

```python
plt.table(rows,cellColours=colors)
plt.axis(False)
```

```python
rows = np.array(rows)
unique_rows = list()
for row in rows[:,1:]:
    add = True
    for temp_row in unique_rows:
        if np.array_equiv(temp_row,row):
            add = False
    if add:
        unique_rows.append(row)
```
