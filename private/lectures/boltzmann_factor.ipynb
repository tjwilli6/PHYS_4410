```python
import numpy as np
from scipy import special
import matplotlib.pyplot as plt
plt.style.use("/Users/tjwilli/presentation.mplstyle")
```

```python
Ns = 1
Nr = 100
q = 500
```

```python
omega_s = []
eng_s = []
for qs in range(0,20,1):
    eng_s.append(qs)
    mult = special.comb(qs + Ns - 1,qs) * special.comb(q-qs + Nr - 1, q-qs)
    omega_s.append(mult)
```

```python
plt.figure(figsize=(8,3))
plt.plot(eng_s,omega_s)
plt.xlabel(r'$q_s$')
plt.ylabel(r'$\Omega_s$')
```
