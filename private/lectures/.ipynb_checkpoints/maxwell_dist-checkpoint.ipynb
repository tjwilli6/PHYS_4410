```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("/Users/tjwilli/jupyter.mplstyle")
```

```python
m = 4.6e-26 #kg, (N2)
k = 1.38e-23
def maxwell_speed_prob(v,T,m):
    prob = (m/(2*np.pi*k*T) * 4*np.pi*v**2 * np.exp(-m*v**2/(2*k*T)))
    return prob
def plot_maxwell(T,m):
    std = np.sqrt(k * T / m)
    v = np.linspace(0,5*std,1000)
    p = maxwell_speed_prob(v,T,m)
    plt.plot(v,p,label='T={:.0f}'.format(T))
    plt.xlabel('v [m/s]')
    plt.ylabel(r'$\frac{N(v)}{N}$')
```

```python
plot_maxwell(293,m)
```

```python
for T in np.logspace(0,4,5):
    plot_maxwell(T,m)
plt.legend()
```
