```python
x = y = np.linspace(-1,1,11)
for ix in x:
    for iy in y:
        if np.isclose(ix,0) and np.isclose(iy,0):
            continue
        r = np.hypot(ix,iy)
        scale = r**2 * 50
        a = plt.arrow(ix,iy,ix/scale,iy/scale,width=0.02)
plt.xlim(-1,1)
plt.ylim(-1,1)
plt.gca().xaxis.set_visible(False)
plt.gca().yaxis.set_visible(False)
plt.gca().set_aspect('equal')
plt.title("1")
plt.savefig("inverse_square_field.pdf")
```

```python
v = np.sqrt(2*dK/m)
print("{:.2E}".format(v))
```
