```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("/Users/tjwilli/jupyter.mplstyle")
```

```python
grades = np.genfromtxt( "exam_grades_anon.csv",delimiter=',' )[1:,1:]
```

```python
grades_pct = grades / 20 * 100
```

```python
def plot_dist(problem_number=-1,curve=0,padright=0):
    if problem_number==-1:
        col = grades_pct.mean(1) + curve
        plt.title("Total")
    else:
        pnum = problem_number - 1
        col = grades_pct[:,pnum] + curve
        plt.title("Problem {}".format(problem_number))
    median = np.percentile(col, 50)
    upper = np.percentile(col, 75)
    lower = np.percentile(col, 25)

    bins=np.linspace(0, 200, 21)
    vals, bins, _ = plt.hist( col,ec='k',bins=bins )
    
    xlow = bins[np.digitize(col.min(),bins)-1]
    xhi = bins[np.digitize(col.max(),bins)]
    ytext =  vals.max() * 1.1
    plt.ylim(0,ytext*1.2)
    plt.axvline(median,c='k',lw=3,ls='--',label='Median={}'.format(median))
    plt.text(median + 1,ytext,"{:.0f}%\n(median)".format(median))
    
    plt.axvline(lower,c='k',lw=3,ls='--')
    plt.text(lower-1,ytext,"{:.0f}%\n".format(lower) + r'($25^\mathrm{{th}}$ percentile)',ha='right')
    
    plt.axvline(upper,c='k',lw=3,ls='--')
    plt.text(upper+1+padright,ytext,"{:.0f}%\n".format(upper) + r'($75^\mathrm{{th}}$ percentile)')
    plt.xlim(xlow,xhi)
    plt.xlabel("Grade [%]")
    plt.ylabel("Frequency")
```

```python
CURVE = 0
plot_dist(8,curve=CURVE,padright=0)
plt.tight_layout()
plt.savefig("exam_p8.pdf")
```
