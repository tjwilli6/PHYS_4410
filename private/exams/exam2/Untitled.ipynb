```python
import matplotlib.pyplot as plt
import numpy as np
import string
```

```python
fig = plt.figure(figsize=(10,8))
x = np.linspace(-10,10,100)
y = -x**2
plt.plot(x,y,c='k',lw=3)

x1 = -5
plt.scatter(x1,-x1**2,c='k',s=200)
plt.text(x1-.5,-x1**2,"A",fontsize=25,ha='right')
plt.gca().set_aspect('auto')


x1 = 5
plt.scatter(x1,-x1**2,c='k',s=200)
plt.text(x1+.5,-x1**2,"B",fontsize=25,ha='left')
plt.gca().set_aspect('auto')
plt.xlabel('X',fontsize=25)
plt.ylabel('Y',fontsize=25)
plt.gca().set_xticklabels([])
plt.gca().set_xticks([])
plt.gca().set_yticklabels([])
plt.gca().set_yticks([])
#plt.gca().xaxis.set_visible(False)
#plt.gca().yaxis.set_visible(False)
plt.savefig("avg_velocity_fig.pdf")
```

```python
theta = np.arange(0,2*np.pi,np.pi/4)
fig = plt.figure(figsize=(10,8))
letters = (l for l in string.ascii_lowercase)
for angle in theta:
    x = np.cos(angle)
    y = np.sin(angle)
    arrow = plt.Arrow(0,0,x,y,width=0.2,color='k')
    plt.text(x*1.15,y*1.15,next(letters),ha="center",va="center",fontsize=35)
    plt.gca().add_artist(arrow)
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
plt.gca().set_aspect('equal')
plt.gca().xaxis.set_visible(False)
plt.gca().yaxis.set_visible(False)
plt.axis("off")
plt.savefig("arrow_choices.pdf")
```

```python
g = 9.8
x = 200
theta = np.pi/6
vi = np.sqrt(g*x/2/np.cos(theta)/np.sin(theta))
tf = 2 * vi / g * np.sin(theta)
```

```python
t = np.linspace(0,tf,1000)
x = vi * np.cos(theta) * t
y = vi * np.sin(theta) * t - 0.5 * g * t**2
```

```python
plt.figure(figsize=(12,10))
plt.plot(x,y,c='k',ls='--')
plt.axhline(0,c='k')
plt.gca().set_aspect('equal')
plt.gca().tick_params(labelsize=20,length=10)
arrow = plt.Arrow(0,0,np.cos(theta) * 20,np.sin(theta) * 20,width=10,zorder=100)
plt.gca().add_artist(arrow)
plt.text(np.cos(theta)*15,np.sin(theta)*2,r'$30^\circ$',fontsize=15,ha='left')
#plt.xlim(-2,201)
#plt.ylim(-1,30)
plt.gca().set_xticks([0,200])
plt.gca().set_yticks([])
plt.xlabel('X',fontsize=25)
plt.ylabel('Y',fontsize=25)
plt.savefig('projectile_motion.pdf')
```
